# 6.  What are the most popular product keywords ?

```python
df.columns
```

```python
df.product_name.unique()
```

```python
from collections import Counter
import pandas as pd


# Combine all product names into a single string
all_keywords = " ".join(df['product_name'])

# Split the string into individual words
keywords_list = all_keywords.split()

# Count the frequency of each keyword
keyword_counts = Counter(keywords_list)

# Get the most popular keywords
most_popular_keywords = keyword_counts.most_common(20)  # Top 5 keywords

print("Most popular product keywords:")
for keyword, count in most_popular_keywords:
    print(f"{keyword}: {count}")
```
