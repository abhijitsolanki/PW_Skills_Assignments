Q 1.  Demonstrate three different methods for creating identical 2D arrays in NumPy. Provide the code for each
method and the final output after each method.

```python
import numpy as np
import pandas as pd
```

```python
import numpy as np

# Method 1: Using np.array()
arr1 = np.array([[1, 2], [3, 4]])
print("Method 1:\n", arr1)

# Method 2: Using np.ones() or np.zeros() and multiplying/scaling
arr2 = np.ones((2, 2)) * np.array([[1, 2], [3, 4]])
print("Method 2:\n", arr2)

# Method 3: Using np.full()
arr3 = np.full((2, 2), 0)
arr3[0, :] = [1, 2]
arr3[1, :] = [3, 4]
print("Method 3:\n", arr3)
```

Q 5.Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array.
perform the following operations:
 a)Extract all even integers from array.
 b)Extract all odd integers from array.

```python
# Create random array of integers between 1 and 10
arr = np.random.randint(1, 11, (5, 6))

# Extract even integers
even_integers = arr[arr % 2 == 0]
print("Even integers:\n", even_integers)

# Extract odd integers
odd_integers = arr[arr % 2 != 0]
print("Odd integers:\n", odd_integers)
```

Q 14. Create a time-series dataset and plot 'Temperature' and 'Humidity':

```python
# Create a time-series DataFrame with 'Date', 'Temperature', and 'Humidity'
date_range = pd.date_range(start='2024-01-01', periods=100)
temperature = np.random.uniform(15, 30, 100)
humidity = np.random.uniform(50, 90, 100)

df = pd.DataFrame({'Date': date_range, 'Temperature': temperature, 'Humidity': humidity})

# Plot with two y-axes
fig, ax1 = plt.subplots()

ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color='tab:red')
ax1.plot(df['Date'], df['Temperature'], color='tab:red')

ax2 = ax1.twinx()
ax2.set_ylabel('Humidity', color='tab:blue')
ax2.plot(df['Date'], df['Humidity'], color='tab:blue')

plt.title('Temperature and Humidity Over Time')
plt.show()
```

Q 15. Plot a histogram with 30 bins and overlay PDF of normal distribution:

```python
# Create a NumPy array of 1000 samples from a normal distribution
data = np.random.normal(0, 1, 1000)

# Plot the histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Overlay the PDF of normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = np.exp(-0.5 * (x ** 2)) / np.sqrt(2 * np.pi)
plt.plot(x, p, 'k', linewidth=2)

plt.title('Histogram with PDF Overlay')
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.show()
```

Q 16. Using Bokeh, plot a line chart of a sine wave function:
python

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

# Create sine wave data
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Create Bokeh plot
p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Add grid lines
p.xgrid.grid_line_color = 'blue'
p.ygrid.grid_line_color = 'blue'

show(p)
```

Q 17. Using Bokeh, generate a bar chart of random categorical data:


```python
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap

# Create random categorical data
categories = ['A', 'B', 'C', 'D']
values = np.random.randint(1, 100, 4)

source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create Bokeh plot
p = figure(x_range=categories, title="Random Categorical Bar Chart", toolbar_location=None)

# Add bars
p.vbar(x='categories', top='values', width=0.9, source=source, line_color='white',
       fill_color=factor_cmap('categories', palette=['#718dbf', '#e84d60', '#ddb7b1', '#c9d9d3'], factors=categories))

# Add hover tooltips
p.add_tools(HoverTool(tooltips=[("Category", "@categories"), ("Value", "@values")]))

show(p)
```

Q 18. Using Plotly, create a basic line plot of random data:

```python
import plotly.graph_objects as go

# Create random data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create Plotly figure
fig = go.Figure()

# Add trace for line plot
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='sin(x)'))

# Set title and axis labels
fig.update_layout(title='Simple Line Plot', xaxis_title='x', yaxis_title='sin(x)')

fig.show()
```

Q 20. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants):

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create random data
x = np.random.randn(100)
y = np.random.randn(100)

# Create a DataFrame for plotting
df = pd.DataFrame({'x': x, 'y': y})

# Determine quadrants
df['quadrant'] = np.where((df['x'] >= 0) & (df['y'] >= 0), 'Q1',
                          np.where((df['x'] < 0) & (df['y'] >= 0), 'Q2',
                                   np.where((df['x'] < 0) & (df['y'] < 0), 'Q3', 'Q4')))

# Create scatter plot with color based on quadrant
sns.scatterplot(data=df, x='x', y='y', hue='quadrant', palette='Set1')

# Add labels and title
plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(title='Quadrant')

# Show the plot
plt.show()
```

Q 21. Create a time-series dataset and plot 'Temperature' and 'Humidity' on the same plot with different y-axes:

```python
# Create a time-series dataset
date_range = pd.date_range(start='2024-01-01', periods=100)
temperature = np.random.uniform(15, 30, 100)
humidity = np.random.uniform(50, 90, 100)

df = pd.DataFrame({'Date': date_range, 'Temperature': temperature, 'Humidity': humidity})

# Plot Temperature and Humidity with different y-axes
fig, ax1 = plt.subplots()

# Plot Temperature
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.tick_params(axis='y', labelcolor='tab:blue')

plt.title('Temperature and Humidity Over Time')
fig.tight_layout()
plt.show()
```
