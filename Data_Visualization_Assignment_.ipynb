Q  5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
num_samples = 500
# Generate three features where two of them are correlated
feature_1 = np.random.normal(0, 1, num_samples)
feature_2 = 2 * feature_1 + np.random.normal(0, 0.5, num_samples)  # correlated with feature_1
feature_3 = np.random.normal(0, 1, num_samples)  # independent feature

# Combine into a DataFrame
data = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3
})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.express as px

# Provided dataset
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),  # Population data
    'GDP': np.random.randint(500, 2000, 5)          # GDP data
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a bubble chart using Plotly Express
fig = px.scatter(df, x='GDP', y='Population', size='Population',
                 text='Country',  # Show country name as text
                 title='Country Population vs GDP',
                 labels={'GDP': 'GDP (in billion USD)', 'Population': 'Population (in millions)'},
                 size_max=60)  # Max bubble size


# Show the plot
fig.show()
```

```python
import plotly.graph_objects as go

# Create the figure
fig = go.Figure()

# Add a scatter plot trace
fig.add_trace(go.Scatter(
    x=df.GDP,
    y=df.Population,
    mode='markers',
    marker=dict(size=1/5 * df['Population']),  # Corrected marker size
    text=df.Country  # Display country names on hover
))

# Update layout with titles and labels
fig.update_layout(
    title='Country Population vs GDP',
    xaxis_title='GDP (in billion USD)',
    yaxis_title='Population (in millions)'
)

# Show the plot
fig.show()
```

Q  2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
import numpy as np

# Generate random data
np.random.seed(42)
n = 50
data = {
    'x': np.random.rand(n) * 100,
    'y': np.random.rand(n) * 100,
    'sizes': np.random.rand(n) * 20 + 5,  # Random sizes between 5 and 25
    'colors': np.random.choice(['red', 'green', 'blue', 'orange'], n)  # Random colors
}

df = pd.DataFrame(data)

# Create a scatter plot
output_notebook()
p_scatter = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y')
p_scatter.scatter('x', 'y', source=df, size='sizes', color='colors', legend_field='colors', fill_alpha=0.6)

# Show the plot
show(p_scatter)
```

Q 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

 fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

 counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Prepare the data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a bar chart
output_notebook()  # Use this for Jupyter Notebooks
p_bar = figure(title="Fruit Counts", x_range=fruits, plot_height=400, toolbar_location=None)
p_bar.vbar(x='fruits', top='counts', width=0.9, source=source, fill_color='blue')

# Customize the axes
p_bar.xgrid.grid_line_color = None
p_bar.axis.minor_tick_line_color = None
p_bar.outline_line_color = None
p_bar.xaxis.axis_label = "Fruits"
p_bar.yaxis.axis_label = "Counts"

# Show the plot
show(p_bar)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Generate random data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a histogram
output_notebook()
p_hist = figure(title="Histogram", x_axis_label='Value', y_axis_label='Count')
p_hist.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='blue', line_color='white')

# Show the plot
show(p_hist)
```

 5. Create a Bokeh heatmap using the provided dataset.

 data_heatmap = np.random.rand(10, 10)

 x = np.linspace(0, 1, 10)

 y = np.linspace(0, 1, 10)

 xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource
from bokeh.palettes import Viridis256

# Generate random heatmap data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Convert to a flat array
data_flat = data_heatmap.flatten()
x_flat = xx.flatten()
y_flat = yy.flatten()

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, value=data_flat))

# Create a heatmap
output_notebook()  # Use this for Jupyter Notebooks
p_heatmap = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', tools="hover")
p_heatmap.xgrid.grid_line_color = None
p_heatmap.ygrid.grid_line_color = None

# Create a color mapper
mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(data_flat), high=max(data_flat))

# Add rectangles for heatmap
p_heatmap.rect(x='x', y='y', width=0.1, height=0.1, source=source, fill_color=mapper, line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
p_heatmap.add_layout(color_bar, 'right')

# Show the plot
show(p_heatmap)
```
